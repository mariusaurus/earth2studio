## plot average error of mean

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def bring_to_common_length(arr, max_val):
    if max_paired_len - len(arr) <= 0:
        return arr[:max_paired_len]
    else:
        return np.pad(arr, (0, max_paired_len - len(arr)), 'constant' , constant_values=0)


def average_lead_time_errors_over_cases(ens_means):
    err_lt = {}
    vars = ['dist', 'msl_err_of_mean', 'wind_speed_err_of_mean']
    for var in vars:
        errs, weights = [], []
        for _, means in ens_means.items():
            errs.append(bring_to_common_length(means['mean'][var], max_paired_len))
            weights.append(bring_to_common_length( means['n_members'], max_paired_len))

        errs = np.stack(errs)
        weights = np.stack(weights)

        # set nans in err to 0 and their associated weights also to 0
        weights[np.isnan(errs)] = 0
        errs[np.isnan(errs)] = 0

        # compute weighted average
        err_lt[var] = np.average(errs, axis=-2, weights=weights)

    return err_lt


def plot_ensemble_mean_errors_over_lead_time(err_lt, lead_times, n_cases,
                               norm_dict = {'msl_err_of_mean': 100, 'wind_speed_err_of_mean': 1, 'dist': 1000},
                               unit_dict = {'msl_err_of_mean': 'hPa', 'wind_speed_err_of_mean': 'm/s', 'dist': 'km'},
                               out_path: str | None = None,
                               max_lt: np.timedelta64 | None = None):

    """
    plot mean absolute error of ensemble mean weighted by matched ensemble size
    """

    # find last index in lead_times that is equal to or below max_lt (xlim doesn't work on timedelta)
    if max_lt:
        max_lt_idx = np.where(lead_times <= max_lt)[0][-1]
    else:
        max_lt_idx = len(lead_times)

    fig, ax = plt.subplots(len(err_lt), figsize=((len(err_lt)+1)*2, 8), sharex=True)

    # Convert lead_times to hours for plotting
    lead_times_hours = lead_times / np.timedelta64(1, 'h')

    for ivar, var in enumerate(err_lt.keys()):
        ax[ivar].plot(lead_times_hours[:max_lt_idx], err_lt[var][:max_lt_idx]/norm_dict[var])
        ax[ivar].set_ylabel(f'{var} [{unit_dict[var]}]', fontsize=12, weight='bold')

        if ivar == len(err_lt) - 1:
            ax[ivar].set_xlabel('lead time [h]', fontsize=12)
        ax[ivar].grid(True)

    sup_title = f'average error of ensemble means over {n_cases} cases \n'
    sup_title += '(weighted by number of matched tracks at lead time)'
    fig.suptitle(sup_title, fontsize=16)

    fig.tight_layout()
    if out_path:
        plt.savefig(os.path.join(out_path, f'error_metrics_of_ens_mean_over_lead_time.png'))

    return


def plot_ensemble_mean_metrics(ens_means, out_path=None):

    err_lt = average_lead_time_errors_over_cases(ens_means)

    plot_ensemble_mean_errors_over_lead_time(err_lt=err_lt,
                                    lead_times=np.arange(max_paired_len)*np.timedelta64(6, 'h'),
                                    n_cases=len(ens_means),
                                    out_path=out_path,
                                    max_lt=np.timedelta64(192, 'h'))

    return

plot_ensemble_mean_metrics(ens_means, out_dir)
```

## scatter plot of maximum intensities

```python
import numpy as np

plt.close('all')
fig, ax = plt.subplots(1, 2, figsize=(10, 8))

for ii, (var, npfun, minmax) in enumerate(zip(['msl', 'wind_speed'], [np.nanmin, np.nanmax], ['min', 'max'])):
    tick_pos, tick_label = [], []
    for case, ext in extremes.items():
        yy = ext[var][minmax+'_pred']
        xx = ext[var][minmax+'_tru']*np.ones_like(yy)
        ax[ii].plot(xx, yy, marker='o', linestyle='None', color='black', alpha=.1)

    legend = False
    for case, ext in extremes.items():
        yy = ext[var][minmax+'_pred']
        if not legend:
            ax[ii].plot(ext[var][minmax+'_tru'], npfun(ens_means[case]['mean'][var]), marker='o', linestyle='None', color='lime', alpha=1, label=f'{minmax} of mean')
            ax[ii].plot(ext[var][minmax+'_tru'], np.nanmean(yy), marker='o', linestyle='None', color='red', alpha=1, label=f'mean of {minmax}')
            legend = True
        else:
            ax[ii].plot(ext[var][minmax+'_tru'], npfun(ens_means[case]['mean'][var]), marker='o', linestyle='None', color='lime', alpha=1)
            ax[ii].plot(ext[var][minmax+'_tru'], np.nanmean(yy), marker='o', linestyle='None', color='red', alpha=1)

        tick_pos.append(ext[var][minmax+'_tru'])
        tick_label.append(case)

    # ax[ii].legend()
    ax[ii].set_aspect('equal')
    ax[ii].legend()
    tick_pos_dflt = ax[ii].get_xticks()
    labels_dflt = [f'{xx:.0f}' for xx in tick_pos_dflt]

    tick_pos = np.append(tick_pos_dflt, tick_pos)
    tick_label = np.append(labels_dflt, tick_label)

    ax[ii].set_xticks(tick_pos)
    ax[ii].set_xticklabels(tick_label, rotation=90, ha='right')

    # 2. Get the limits of the axes
    low_x, high_x = ax[ii].get_xlim()
    low_y, high_y = ax[ii].get_ylim()

    # 3. Calculate the range for the diagonal line
    # This ensures the line is drawn only within the plotted area
    low = max(low_x, low_y)
    high = min(high_x, high_y)
    ax[ii].set_xlim(low, high)
    ax[ii].set_ylim(low, high)

    # 4. Plot the diagonal line
    ax[ii].plot([low, high], [low, high], ls='--', color='k', label='f(x) = x', zorder=0, alpha=.6)

fig.tight_layout()
```
